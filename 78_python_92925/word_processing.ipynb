```python
import re, csv
positive_words = []
negative_words = []
print("Positive words in IV-4")
with open('iv4.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if row['Positiv']:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Entry']))
            print(word)
            positive_words.append(word)
```

```python
print("Negativ words in IV-4")
with open('iv4.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if row['Negativ']:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Entry']))
            print(word)
            negative_words.append(word)
```

```python
print("Positive words in MD")
with open('md2020.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if int(row['Positive'])>0:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Word']).upper())
            print(word)
            positive_words.append(word)
            
```

```python
print("Negative words in MD")
with open('md2020.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if int(row['Negative'])>0:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Word']).upper())
            print(word)
            positive_words.append(word)
```
