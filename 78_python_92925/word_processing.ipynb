```python
import re, csv
positive_words = []
negative_words = []
print("Positive words in IV-4")
with open('iv4.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if row['Positiv']:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Entry']))
            print(word)
            positive_words.append(word)
```

```python
print("Negativ words in IV-4")
with open('iv4.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if row['Negativ']:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Entry']))
            print(word)
            negative_words.append(word)
```

```python
print("Positive words in MD")
with open('md2020.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if int(row['Positive'])>0:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Word']).upper())
            print(word)
            positive_words.append(word)
            
```

```python
print("Negative words in MD")
with open('md2020.csv' , 'r') as f:
    csvreader=csv.DictReader(f)
    for row in csvreader:
        if int(row['Negative'])>0:
            word = (re.sub(r'[^a-zA-Z\s]', '', row['Word']).upper())
            print(word)
            positive_words.append(word)
```

```python
text = "I thought they were pretty nice at first. Sound quality is good. They charged quick in the case. But transparency mode amplified sounds like crazy. At the gym someone was moving weights, the cling of them hitting was so loud it hurt my ears. When I took the airpods out it was much quieter in real life. Even in noise cancelling mode they pick up voices and then they end up sounding robotic. There's a white noise in the background too. It's like a little static, annoying as hell. Just now I started hearing a popping noise in the right bud. It kept happening even when the music was off. When I turned to transparency mode that popping became a buzz."
```

```python
def Convert(string):
    li = list(string.split(" "))
    return li
```

```python
word1 = (Convert(text))
print(word1)
```

```python
for i in range(len(word1)):
    word1[i] = word1[i].lower()
print(word1)
```

```python
found_positive_words = []
found_negative_words = []
```

```python
list3 = set(word1)&set(positive_words)

list4 = sorted(list3, key = lambda k : word1.index(k))

found_positive_words.append(list4)
```

```python
print(found_positive_words)
print("Positive Words", len(found_positive_words))
```

```python
list3 = set(word1)&set(negative_words)

list4 = sorted(list3, key = lambda k : word1.index(k))

found_negative_words.append(list4)
```

```python

print(found_negative_words)
print("Negative Words", len(found_negative_words))
```

```python
def colourise(colour, text):
    if colour == "black":
        return "\033[1;30m" + str(text) + "\033[1;m"
    if colour == "red":
        return "\033[1;31m" + str(text) + "\033[1;m"
    if colour == "green":
        return "\033[1;32m" + str(text) + "\033[1;m"
    if colour == "yellow":
        return "\033[1;33m" + str(text) + "\033[1;m"
    if colour == "blue":
        return "\033[1;34m" + str(text) + "\033[1;m"
    if colour == "magenta":
        return "\033[1;35m" + str(text) + "\033[1;m"
    if colour == "cyan":
        return "\033[1;36m" + str(text) + "\033[1;m"
    if colour == "gray":
        return "\033[1;37m" + str(text) + "\033[1;m"
    return str(text)
```

```python
def highlight(colour, text):
    if colour == "black":
        return "\033[1;40m" + str(text) + "\033[1;m"
    if colour == "red":
        return "\033[1;41m" + str(text) + "\033[1;m"
    if colour == "green":
        return "\033[1;42m" + str(text) + "\033[1;m"
    if colour == "yellow":
        return "\033[1;43m" + str(text) + "\033[1;m"
    if colour == "blue":
        return "\033[1;44m" + str(text) + "\033[1;m"
    if colour == "magenta":
        return "\033[1;45m" + str(text) + "\033[1;m"
    if colour == "cyan":
        return "\033[1;46m" + str(text) + "\033[1;m"
    if colour == "gray":
        return "\033[1;47m" + str(text) + "\033[1;m"
    return str(text)
```

```python
print(highlight("red", (colourise("blue", found_negative_words))))
```
